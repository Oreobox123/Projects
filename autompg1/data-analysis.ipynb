```python
import matplotlib.pyplot as plt

# Scatter plot: horsepower vs mpg
plt.figure(figsize=(10, 6))
plt.scatter(df['horsepower'], df['mpg'], color='teal', alpha=0.6)

plt.title('Horsepower vs MPG')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')
plt.grid(True)
plt.show()
```

```python
# Replace '?' with NaN
df['horsepower'] = df['horsepower'].replace('?', pd.NA)

# Convert horsepower to numeric (coerces errors to NaN)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows with missing values (only if you’re okay with removing them)
df.dropna(subset=['horsepower', 'mpg'], inplace=True)
df[['horsepower', 'mpg']].corr()
```
